```python
import pysrt
from googletrans import Translator
import time

# Load the input SRT file
input_file = "P7.srt"
subs = pysrt.open(input_file)

# Initialize the Google Translate API client
translator = Translator()

# Process each subtitle in the SRT file
with open("output.srt", "w", encoding="utf-8") as f:
    for i in range(len(subs)):
        # Get the start and end times
        start_time = subs[i].start.to_time().strftime('%H:%M:%S,%f')[:-3]
        end_time = subs[i].end.to_time().strftime('%H:%M:%S,%f')[:-3]

        # Get the English subtitle text
        sub_text = subs[i].text

        # Translate the subtitle text to Persian
        try:
            translation = translator.translate(sub_text, dest='fa')
            persian_sub_text = translation.text

            # Add a delay between translation requests to avoid hitting rate limits

        except Exception as e:
            try:
                time.sleep(2)
                translation = translator.translate(sub_text, dest='fa')
                persian_sub_text = translation.text
            
            
            except Exception as e:
                try:
                    time.sleep(4)
                    translation = translator.translate(sub_text, dest='fa')
                    persian_sub_text = translation.text
                    
            
            
                except Exception as e: 
                    try:
                        time.sleep(8)
                        translation = translator.translate(sub_text, dest='fa')
                        persian_sub_text = translation.text

            
                    except Exception as e: 
                
                        print(f"Error translating subtitle {i+1}: {e}")
                        persian_sub_text = sub_text

        # Combine the time info and translated subtitle text
        output_text = f"{i+1}\n{start_time} --> {end_time}\n{persian_sub_text}\n\n"

        # Write the output text to a new SRT file
        f.write(output_text)

        # Print the output text to console
        print(output_text)

# Read the contents of the output SRT file and print it to console
with open("output.srt", "r", encoding="utf-8") as f:
    srt_contents = f.read()
    print(srt_contents)
```
